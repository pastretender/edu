# **Parameters**
**nc**: Number of classes that YOLOv5 should detect.

**depth_multiple**: A scaling factor for the model depth, allowing different versions (small, medium, large) of YOLOv5 to have varying depths.

**width_multiple**: A scaling factor for the model width (number of channels), adjusting the number of channels in each layer.

**Anchors**:
Anchors are predefined bounding boxes of varying sizes that help the model detect objects of different scales:

P3/8: Anchors for smaller objects.
P4/16: Anchors for medium-sized objects.
P5/32: Anchors for larger objects.

**Backbone**:
The backbone is the feature extractor in YOLOv5, and it’s responsible for extracting spatial information from the input image. It includes:

**Focus layer:** Used to reduce the input image resolution by concatenating feature maps, which allows the model to capture rich spatial information.

**Conv layers:** These are convolutional layers, with a stride of 2, used for down-sampling the spatial dimensions.

**BottleneckCSP:** CSP (Cross Stage Partial) layers help improve feature propagation and reduce computational load.

**SPP (Spatial Pyramid Pooling):** Used at the end of the backbone to create fixed-size feature maps, enhancing the model's ability to detect objects at different scales.

**Upsample layers:** Used to increase the spatial resolution of feature maps from the lower levels to match the higher levels, enhancing detail retention for smaller objects.

**Concat layers:** Combine feature maps from different stages of the backbone to improve the detection's accuracy by leveraging multiscale information.

**Detect layer:** Uses the three feature scales (P3, P4, and P5) to predict the bounding boxes and class probabilities for small, medium, and large objects.

```
%%writetemplate /content/yolov5/models/custom_yolov5s.yaml

# parameters
nc: {num_classes}  # number of classes
depth_multiple: 0.33  # model depth multiple
width_multiple: 0.50  # layer channel multiple

# anchors
anchors:
  - [10,13, 16,30, 33,23]  # P3/8
  - [30,61, 62,45, 59,119]  # P4/16
  - [116,90, 156,198, 373,326]  # P5/32

# YOLOv5 backbone
backbone:
  # [from, number, module, args]
  [[-1, 1, Focus, [64, 3]],  # 0-P1/2
   [-1, 1, Conv, [128, 3, 2]],  # 1-P2/4
   [-1, 3, BottleneckCSP, [128]],
   [-1, 1, Conv, [256, 3, 2]],  # 3-P3/8
   [-1, 9, BottleneckCSP, [256]],
   [-1, 1, Conv, [512, 3, 2]],  # 5-P4/16
   [-1, 9, BottleneckCSP, [512]],
   [-1, 1, Conv, [1024, 3, 2]],  # 7-P5/32
   [-1, 1, SPP, [1024, [5, 9, 13]]],
   [-1, 3, BottleneckCSP, [1024, False]],  # 9
  ]

# YOLOv5 head
head:
  [[-1, 1, Conv, [512, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 6], 1, Concat, [1]],  # cat backbone P4
   [-1, 3, BottleneckCSP, [512, False]],  # 13

   [-1, 1, Conv, [256, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 4], 1, Concat, [1]],  # cat backbone P3
   [-1, 3, BottleneckCSP, [256, False]],  # 17 (P3/8-small)

   [-1, 1, Conv, [256, 3, 2]],
   [[-1, 14], 1, Concat, [1]],  # cat head P4
   [-1, 3, BottleneckCSP, [512, False]],  # 20 (P4/16-medium)

   [-1, 1, Conv, [512, 3, 2]],
   [[-1, 10], 1, Concat, [1]],  # cat head P5
   [-1, 3, BottleneckCSP, [1024, False]],  # 23 (P5/32-large)

   [[17, 20, 23], 1, Detect, [nc, anchors]],  # Detect(P3, P4, P5)
  ]
```

# Train Custom YOLOv5 Detector



Here, we are able to pass a number of arguments:
- **img:** define input image size
- **batch:** determine batch size
- **epochs:** define the number of training epochs. (Note: often, 3000+ are common here!)
- **data:** set the path to our yaml file
- **cfg:** specify our model configuration
- **weights:** specify a custom path to weights. (Note: you can download weights from the Ultralytics Google Drive [folder](https://drive.google.com/open?id=1Drs_Aiu7xx6S-ix95f9kNsA6ueKRpN2J))
- **name:** result names
- **nosave:** only save the final checkpoint
- **cache:** cache images for faster training
